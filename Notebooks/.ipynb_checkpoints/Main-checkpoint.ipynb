# Denoise

```python
from image_restoration_web_app.Dataset.Div2k import Div2kDataset
from image_restoration_web_app.Dataset.SIDD import SIDDDataset
from torch.utils.data import Subset

sidd_path = '/content/drive/MyDrive/Training_model/dataset/SIDD/Data'
div2k_path = '/content/drive/MyDrive/Training_model/dataset/train'

noise_levels = [0.5, 1, 1.5, 2]
patch_size = 512

sidd_train = SIDDDataset(sidd_path, patch_size=patch_size)
sidd_dataloader_train = DataLoader(sidd_train, shuffle=True)
sidd_val = SIDDDataset(sidd_path, patch_size=patch_size, train=False)
sidd_dataloader_val = DataLoader(sidd_val, shuffle=True)

div2k_train = Div2kDataset(div2k_path, noise_levels=noise_levels, patch_size=patch_size)
div2k_train_subset_indices = list(range(len(sidd_train)))
div2k_train_subset = Subset(div2k_train, div2k_train_subset_indices)
div2k_dataloader_train = DataLoader(div2k_train_subset, shuffle=True)


div2k_val = Div2kDataset(div2k_path, noise_levels=noise_levels, patch_size=patch_size, train=False)
div2k_val_subset_indices = list(range(len(sidd_val)))
div2k_val_subset = Subset(div2k_val, div2k_val_subset_indices)
div2k_dataloader_val = DataLoader(div2k_val, shuffle=True)
```

```python
print("sidd_train length => ", len(sidd_train))
print("sidd_val length => ", len(sidd_val))
print("div2k_train length => ", len(div2k_train_subset))
print("div2k_val length => ", len(div2k_val_subset))
```

## Train

```python
from image_restoration_web_app.src.denoise_train import train

epoch = 49
folder = "/content/drive/MyDrive/Training_model"
checkpoint_path = f"{folder}/Denoise_epoch_{epoch}.pth"
excel_file = "/content/drive/MyDrive/Training_model/TrainingProgress/training_results_Denoise.xlsx"
progress_file = f"{folder}/TrainingProgress/training_progress_Denoise_{epoch+1}.txt"

train(sidd_dataloader_train, div2k_dataloader_train, folder, checkpoint_path, excel_file, progress_file, device="cuda")
```
